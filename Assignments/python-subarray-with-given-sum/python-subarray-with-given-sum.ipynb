## Solution

Test case:

```python
arr1 = [1, 7, 4, 2, 1, 3, 11, 5]
target1 = 10
```

Brute force $O(n^3)$:

```python
def subarray_sum1(arr, target):
    n = len(arr)
    for i in range(n):
        for j in range(i, n+1):
            if sum(arr[i:j]) == target:
                return i, j
    return None, None
```

```python
i, j = subarray_sum1(arr1, target1)
i, j, arr1[i:j]
```

Better brute force $O(n^2)$:

```python
def subarray_sum2(arr, target):
    n = len(arr)
    for i in range(n):
        s = 0
        for j in range(i, n+1):
            if s == target:
                return i,j
            s += arr[j]
    return None, None
```

```python
i, j = subarray_sum2(arr1, target1)
i, j, arr1[i:j]
```

Greedy algorithm $O(n)$:

```python
def subarray_sum3(arr, target):
    n = len(arr)
    i, j, s = 0, 0, 0
    while i < n and j <= n:
        if s == target:
            return i, j
        elif s < target:
            s += arr[j]
            j += 1
        elif s > target:
            s -= arr[i]
            i += 1
    return None, None
        
```

```python
i, j = subarray_sum3(arr1, target1)
i, j, arr1[i:j]
```
